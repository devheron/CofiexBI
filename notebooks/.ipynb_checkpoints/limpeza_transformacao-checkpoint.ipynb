```python
from pathlib import Path

import numpy as np
import pandas as pd
```

```python
BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

ARQUIVO_BRUTO = BASE_DIR / "dados_brutos" / "dados_2026-04-14.xlsx"

df_raw = pd.read_excel(ARQUIVO_BRUTO, sheet_name="Dados")
df_tratado = df_raw.copy()

print(f"df_raw: {df_raw.shape}")
print(f"df_tratado: {df_tratado.shape}")
```

```python
df_tratado["de_tipo_operacao_padronizado"] = (
    df_tratado["de_tipo_operacao"]
    .astype("string")
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
mapa_tipo_operacao = {
    "Contribuição Financeira não Reembolsável GEF": "Contribuição financeira não reembolsável",
    "Contribuição financeira não reembolsável": "Contribuição financeira não reembolsável"
}

df_tratado["de_tipo_operacao_padronizado"] = (
    df_tratado["de_tipo_operacao_padronizado"]
    .replace(mapa_tipo_operacao)
)
```

```python
df_tratado["de_tipo_operacao_padronizado"].value_counts(dropna=False)
```

```python
colunas_texto = df_tratado.select_dtypes(include=["object", "string"]).columns

colunas_texto
```

```python
for coluna in colunas_texto:
    df_tratado[coluna] = (
        df_tratado[coluna]
        .astype("string")
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )
```

```python
df_tratado["de_fase_padronizado"] = (
    df_tratado["de_fase"]
    .astype("string")
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
mapa_fase = {}

df_tratado["de_fase_padronizado"] = (
    df_tratado["de_fase_padronizado"]
    .replace(mapa_fase)
)
```

```python
df_tratado["de_fase_padronizado"].value_counts(dropna=False)
```

```python
df_tratado["de_esfera_padronizado"] = (
    df_tratado["de_esfera"]
    .astype("string")
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
df_tratado["de_esfera_padronizado"] = (
    df_tratado["de_esfera_padronizado"]
    .fillna("Não informado")
)
```

```python
df_tratado["de_esfera_padronizado"].value_counts(dropna=False)
```

```python
df_tratado["nm_regiao_padronizado"] = (
    df_tratado["nm_regiao"]
    .astype("string")
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
mapa_regiao = {}

df_tratado["nm_regiao_padronizado"] = (
    df_tratado["nm_regiao_padronizado"]
    .replace(mapa_regiao)
)
```

```python
df_tratado["nm_regiao_padronizado"].value_counts(dropna=False)
```

```python
df_tratado["sg_uf_padronizado"] = (
    df_tratado["sg_uf"]
    .astype("string")
    .str.strip()
    .str.upper()
    .str.replace(r"\s+", "", regex=True)
)
```

```python
df_tratado["sg_uf_padronizado"] = (
    df_tratado["sg_uf_padronizado"]
    .fillna("NAO_INFORMADO")
)
```

```python
df_tratado["sg_fonte_padronizado"] = (
    df_tratado["sg_fonte"]
    .astype("string")
    .str.strip()
    .str.upper()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
df_tratado["sg_fonte_padronizado"] = (
    df_tratado["sg_fonte_padronizado"]
    .fillna("NAO_INFORMADO")
)
```

```python
df_tratado["sg_uf_padronizado"].value_counts(dropna=False).head(30)
```

```python
df_tratado["sg_fonte_padronizado"].value_counts(dropna=False).head(30)
```

```python
print(
    "Categorias sg_uf antes:",
    df_tratado["sg_uf"].nunique(dropna=False)
)

print(
    "Categorias sg_uf depois:",
    df_tratado["sg_uf_padronizado"].nunique(dropna=False)
)

print(
    "Categorias sg_fonte antes:",
    df_tratado["sg_fonte"].nunique(dropna=False)
)

print(
    "Categorias sg_fonte depois:",
    df_tratado["sg_fonte_padronizado"].nunique(dropna=False)
)
```

```python
# Padronização de sg_fonte
df_tratado["sg_fonte_padronizado"] = (
    df_tratado["sg_fonte"]
    .astype("string")
    .str.strip()
    .str.upper()
    .str.replace(r"\s+", " ", regex=True)
    .fillna("NAO_INFORMADO")
)

# Mapa de correções na coluna padronizada
mapa_sg_fonte = {
    "BID/ICSF": "BID/ICSTF",
}

df_tratado["sg_fonte_padronizado"] = (
    df_tratado["sg_fonte_padronizado"]
    .replace(mapa_sg_fonte)
)

# SECI: preencher sigla ausente com base na descrição
cond_seci = (
    (df_tratado["sg_fonte_padronizado"] == "NAO_INFORMADO")
    & df_tratado["de_fonte"].str.contains("SECI", na=False)
)
df_tratado.loc[cond_seci, "sg_fonte_padronizado"] = "SECI"
```

```python
df_tratado["sg_fonte_padronizado"].value_counts(dropna=False)
```

```python
# Tratar campos textuais vazios como nulos reais
colunas_texto = df_tratado.select_dtypes(include=["object", "string"]).columns

for coluna in colunas_texto:
    df_tratado[coluna] = (
        df_tratado[coluna]
        .replace(r"^\s*$", pd.NA, regex=True)
    )
```

```python
df_tratado[colunas_texto].isna().sum()
```

```python
# Criar flags para campos ausentes importantes
df_tratado["flag_sem_fonte"] = df_tratado["sg_fonte"].isna()
df_tratado["flag_sem_financiamento"] = df_tratado["vl_financiamento_dolar"].isna()
df_tratado["flag_sem_contrapartida"] = df_tratado["vl_contrapartida_dolar"].isna()
df_tratado["flag_sem_primeira_cofiex"] = df_tratado["dt_primeira_cofiex"].isna()
df_tratado["flag_sem_assinatura"] = df_tratado["dt_assinatura"].isna()
```

```python
df_tratado[["flag_sem_fonte", "flag_sem_financiamento", "flag_sem_contrapartida",
            "flag_sem_primeira_cofiex", "flag_sem_assinatura"]].sum()
```

```python
# Preservar colunas originais de data antes da conversão
colunas_data = [
    "dt_ultimo_recebimento",
    "dt_primeira_cofiex",
    "dt_validade_recomendacao",
    "dt_reuniao_negociacao",
    "dt_publicacao_dou",
    "dt_assinatura",
    "dt_ultimo_desembolso_original",
    "dt_ultimo_desembolso_vigente",
]

for coluna in colunas_data:
    df_tratado[f"{coluna}_original"] = df_tratado[coluna].copy()
```

```python
for coluna in colunas_data:
    df_tratado[coluna] = pd.to_datetime(df_tratado[coluna], errors="coerce")
```
