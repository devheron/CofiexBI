```python
import pandas as pd

# caminho do arquivo
file_path = '../data/raw/dados_2025-02-05.xlsx'

#df = pd.read_csv(file_path)
df = pd.read_excel(file_path)

df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
# ========= TRANSFORM FINAL =========

# remover duplicados
df = df.drop_duplicates()

# NUMÉRICOS
df['VlDolar'] = pd.to_numeric(df['VlDolar'], errors='coerce')
df['CpVlDolar'] = pd.to_numeric(df['CpVlDolar'], errors='coerce')

# DATAS
date_columns = [
    'dtRecebimento',
    'dtValidadeRecomendacao',
    'dtReuniaoNegociacao',
    'dtAprovDiretoria',
    'dtAprovacaoSenado',
    'dtAssinatura',
    'dtPrimeiraAmortizacao',
    'dtUltimaAmortizacao',
    'dtUltimoDesembolsoOriginal',
    'dtUltimoDesembolsoVigente'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)

# TEXTOS
text_columns = [
    'nmProjeto',
    'nmSujeito',
    'nmSetor',
    'Regiao',
    'TipoFinanciamento'
]

for col in text_columns:
    df[col] = df[col].str.strip().str.upper()

# FEATURE ENGINEERING
df['ano_recebimento'] = df['dtRecebimento'].dt.year
print("Arquvio Transformado!")
```

```python
# Save Load Final

output_path = '../data/processed/cofiex_dadostratados.csv'

df.to_csv(output_path, index=False)
print("Tudo certo Arquivo transportado para data/processed!")
```
