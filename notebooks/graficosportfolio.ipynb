```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/processed/cofiex_dadostratados.csv')

df.head()
df.info()
```

```python
#Valor total por regiao
df.groupby('Regiao')['VlDolar'].sum().sort_values(ascending=False)
```

```python
# Grafico por região
import matplotlib.pyplot as plt

df.groupby('Regiao')['VlDolar'].sum().sort_values().plot(kind='barh')

plt.title('Valor Total por Região')
plt.xlabel('Valor em Dólar')
plt.ylabel('Região')
plt.show()
```

```python
# Projetos por ano
df['ano_recebimento'].value_counts().sort_index()
```

```python
#Grafico temporal
df['ano_recebimento'].value_counts().sort_index().plot()

plt.title('Projetos por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade')
plt.show()
```

```python
# Top10 projetos
df[['nmProjeto', 'VlDolar']].sort_values(by='VlDolar', ascending=False).head(10)
```

```python
# Projetos por região
df['Regiao'].value_counts().plot(kind='bar')
plt.title('Projetos por Região')
plt.show()
```

```python
# Top setores
df['nmSetor'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Setores')
plt.xticks(rotation=45)
plt.show()
```

```python
# Evolução no tempo
df['ano_recebimento'].value_counts().sort_index().plot()
plt.title('Projetos por Ano')
plt.show()
```

```python
# Valores
df['VlDolar'].plot(kind='hist', bins=30)
plt.title('Distribuição de Valores (USD)')
plt.show()
```

```python
# Tipo financiamento
df['TipoFinanciamento'].value_counts().plot(kind='bar')
plt.title('Tipos de Financiamento')
plt.show()
```
