## Validacao das datas convertidas

Apos `pd.to_datetime(..., errors="coerce")`, conferir minimos, maximos e datas fora de faixa plausivel. Datas sentinela ou futuras extremas viram flag, nunca exclusao automatica.

```python
# Resumo das datas apos conversao
resumo_datas = pd.DataFrame({
    "min": [df_tratado[c].min() for c in colunas_data],
    "max": [df_tratado[c].max() for c in colunas_data],
    "nulos": [df_tratado[c].isna().sum() for c in colunas_data],
}, index=colunas_data)

resumo_datas
```

```python
# Datas fora de faixa plausivel (sentinela / futuras extremas)
# Faixa aceita: 1940-01-01 ate 2035-12-31 (horizonte razoavel de desembolso)
LIMITE_INFERIOR = pd.Timestamp("1940-01-01")
LIMITE_SUPERIOR = pd.Timestamp("2035-12-31")

for coluna in colunas_data:
    fora = (
        df_tratado[coluna].notna()
        & ((df_tratado[coluna] < LIMITE_INFERIOR) | (df_tratado[coluna] > LIMITE_SUPERIOR))
    )
    print(f"{coluna}: {fora.sum()} fora de faixa")

# Flag de data sentinela / fora de faixa (qualquer coluna de data)
mask_data_extrema = pd.Series(False, index=df_tratado.index)
for coluna in colunas_data:
    mask_data_extrema |= (
        df_tratado[coluna].notna()
        & ((df_tratado[coluna] < LIMITE_INFERIOR) | (df_tratado[coluna] > LIMITE_SUPERIOR))
    )

df_tratado["flag_data_extrema"] = mask_data_extrema
print("Total registros com data extrema:", df_tratado["flag_data_extrema"].sum())
```

## Flag para outliers financeiros

Criterio adotado (documentado): valor acima do percentil 99 em financiamento OU contrapartida, OU financiamento nulo/zero com contrapartida muito alta. Valores preservados, apenas sinalizados.

```python
p99_fin = df_tratado["vl_financiamento_dolar"].quantile(0.99)
p99_contra = df_tratado["vl_contrapartida_dolar"].quantile(0.99)

print("percentil 99 financiamento:", p99_fin)
print("percentil 99 contrapartida:", p99_contra)

df_tratado["flag_outlier_financeiro"] = (
    (df_tratado["vl_financiamento_dolar"] > p99_fin)
    | (df_tratado["vl_contrapartida_dolar"] > p99_contra)
    | (
        df_tratado["flag_financiamento_nulo_zero"]
        & (df_tratado["vl_contrapartida_dolar"] > p99_contra)
    )
)

print("registros marcados como outlier financeiro:", df_tratado["flag_outlier_financeiro"].sum())
```

## Tratar dt_ultimo_recebimento igual a 1940-01-01 como placeholder

A data `1940-01-01` aparece 382 vezes apenas em `dt_ultimo_recebimento`. Sem esse placeholder, a data minima real e 1989. Tratada como placeholder: vira `NaT` e e marcada por flag. O original ja foi preservado em `dt_ultimo_recebimento_original`.

```python
PLACEHOLDER_RECEBIMENTO = pd.Timestamp("1940-01-01")

mask_placeholder = df_tratado["dt_ultimo_recebimento"] == PLACEHOLDER_RECEBIMENTO
df_tratado["flag_recebimento_placeholder"] = mask_placeholder

print("registros placeholder 1940-01-01:", mask_placeholder.sum())

df_tratado.loc[mask_placeholder, "dt_ultimo_recebimento"] = pd.NaT

print("min recebimento apos tratamento:", df_tratado["dt_ultimo_recebimento"].min())
print("max recebimento apos tratamento:", df_tratado["dt_ultimo_recebimento"].max())
```

## Tratar datas extremas como sentinela (9999-12-31, 4201-01-01, 2225-01-01 e outras)

Datas claramente impossiveis (`9999-12-31`, `4201-01-01`, `2228-03-29`, `2225-01-01`, `2100-01-01`) sao sentinela/erro. Viram `NaT` e sao marcadas por `flag_data_sentinela`. Datas futuras plausiveis (ate 2052) sao mantidas, pois sao previsoes de desembolso aceitaveis.

```python
DATAS_SENTINELA = [
    pd.Timestamp("9999-12-31"),
    pd.Timestamp("4201-01-01"),
    pd.Timestamp("2228-03-29"),
    pd.Timestamp("2225-01-01"),
    pd.Timestamp("2100-01-01"),
]

mask_sentinela = pd.Series(False, index=df_tratado.index)

for coluna in colunas_data:
    m = df_tratado[coluna].isin(DATAS_SENTINELA)
    if m.sum():
        print(f"{coluna}: {m.sum()} sentinela")
    mask_sentinela |= m
    df_tratado.loc[m, coluna] = pd.NaT

df_tratado["flag_data_sentinela"] = mask_sentinela
print("total registros com data sentinela:", mask_sentinela.sum())
```

```python
# Conferencia dos ranges apos tratamento de placeholder e sentinela
resumo_datas_transf = pd.DataFrame({
    "min": [df_tratado[c].min() for c in colunas_data],
    "max": [df_tratado[c].max() for c in colunas_data],
    "nulos": [df_tratado[c].isna().sum() for c in colunas_data],
}, index=colunas_data)

resumo_datas_transf
```

## Criar valor_total_dolar

`valor_total_dolar = vl_financiamento_dolar + vl_contrapartida_dolar`.

Regra de nulo (documentada): soma tratando ausente como 0, **mas** retorna `NaN` quando ambos sao nulos. Assim um registro sem nenhum valor financeiro nao vira 0 artificial.

```python
fin = df_tratado["vl_financiamento_dolar"]
contra = df_tratado["vl_contrapartida_dolar"]

df_tratado["valor_total_dolar"] = fin.fillna(0) + contra.fillna(0)

ambos_nulos = fin.isna() & contra.isna()
df_tratado.loc[ambos_nulos, "valor_total_dolar"] = np.nan

print("ambos nulos (valor_total NaN):", ambos_nulos.sum())
print(df_tratado["valor_total_dolar"].describe())
```

## Criar percentual_contrapartida

`percentual_contrapartida = vl_contrapartida_dolar / valor_total_dolar`.

Protecao contra divisao por zero e nulo: resultado fica `NaN` quando `valor_total_dolar` eh 0 ou nulo. Expresso em fracao (0 a 1).

```python
total = df_tratado["valor_total_dolar"]

df_tratado["percentual_contrapartida"] = np.where(
    total.notna() & (total != 0),
    df_tratado["vl_contrapartida_dolar"].fillna(0) / total,
    np.nan,
)

print("nulos percentual:", df_tratado["percentual_contrapartida"].isna().sum())
print(df_tratado["percentual_contrapartida"].describe())
```

## Criar prazo_recebimento_cofiex

Dias entre `dt_ultimo_recebimento` e `dt_primeira_cofiex`. Calculado apenas quando ambas as datas sao validas. Valor negativo indica inconsistencia cronologica (sera tratado por flag em validacao).

```python
df_tratado["prazo_recebimento_cofiex"] = (
    df_tratado["dt_primeira_cofiex"] - df_tratado["dt_ultimo_recebimento"]
).dt.days

print("validos:", df_tratado["prazo_recebimento_cofiex"].notna().sum())
print("negativos:", (df_tratado["prazo_recebimento_cofiex"] < 0).sum())
print(df_tratado["prazo_recebimento_cofiex"].describe())
```

## Criar prazo_cofiex_assinatura

Dias entre `dt_primeira_cofiex` e `dt_assinatura`. Apenas datas validas.

```python
df_tratado["prazo_cofiex_assinatura"] = (
    df_tratado["dt_assinatura"] - df_tratado["dt_primeira_cofiex"]
).dt.days

print("validos:", df_tratado["prazo_cofiex_assinatura"].notna().sum())
print("negativos:", (df_tratado["prazo_cofiex_assinatura"] < 0).sum())
print(df_tratado["prazo_cofiex_assinatura"].describe())
```

## Criar prazo_assinatura_desembolso

Dias entre `dt_assinatura` e `dt_ultimo_desembolso_vigente`. Apenas datas validas.

```python
df_tratado["prazo_assinatura_desembolso"] = (
    df_tratado["dt_ultimo_desembolso_vigente"] - df_tratado["dt_assinatura"]
).dt.days

print("validos:", df_tratado["prazo_assinatura_desembolso"].notna().sum())
print("negativos:", (df_tratado["prazo_assinatura_desembolso"] < 0).sum())
print(df_tratado["prazo_assinatura_desembolso"].describe())
```
