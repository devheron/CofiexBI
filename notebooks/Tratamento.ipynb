```python
import pandas as pd

df_tratado = pd.read_csv('../data/processed/cofiex_dadostratados_validade.csv')

df_tratado.head()
```

```python
import pandas as pd
import numpy as np

# Tipos esperados por coluna
tipos_esperados = {
    'cdCartaConsulta': 'int',
    'nmProjeto': 'string',
    'sgProjeto': 'string',
    'nmSetor': 'string',
    'nmSujeito': 'string',
    'Regiao': 'string',
    'nmUnidadeFederal': 'string',
    'nmAbrangenciaNacional': 'string',
    'TipoFinanciamento': 'string',
    'sgFonte': 'string',
    'VlDolar': 'float',
    'CpVlDolar': 'float',
    'sgMoeda': 'string',
    'deFase': 'string',
    'dtRecebimento': 'data',
    'dtValidadeRecomendacao': 'data',
    'dtReuniaoNegociacao': 'data',
    'dtAprovDiretoria': 'data',
    'dtAprovacaoSenado': 'data',
    'nuResolucaoSenado': 'string',
    'dtAssinatura': 'data',
    'nuoperacao': 'string',
    'dtPrimeiraAmortizacao': 'data',
    'dtUltimaAmortizacao': 'data',
    'dtUltimoDesembolsoOriginal': 'data',
    'dtUltimoDesembolsoVigente': 'data'
}
```

```python
def mascara_valida(serie, tipo):
    if tipo == 'string':
        return serie.notna() & (serie.astype(str).str.strip() != '') & (serie.astype(str).str.lower() != 'nan')

    if tipo == 'int':
        convertido = pd.to_numeric(serie, errors='coerce')
        return convertido.notna() & (convertido % 1 == 0)

    if tipo == 'float':
        convertido = pd.to_numeric(serie, errors='coerce')
        return convertido.notna()

    if tipo == 'data':
        convertido = pd.to_datetime(serie, errors='coerce')
        return convertido.notna()

    return serie.notna()
```

```python
resumo = []

for coluna, tipo in tipos_esperados.items():
    if coluna not in df_tratado.columns:
        resumo.append({
            'coluna': coluna,
            'tipo_esperado': tipo,
            'total': len(df_tratado),
            'validos': 0,
            'invalidos_ou_nulos': len(df_tratado),
            'percentual_valido': 0,
            'percentual_invalido': 100
        })
        continue

    validos = mascara_valida(df_tratado[coluna], tipo)
    qtd_validos = validos.sum()
    qtd_invalidos = (~validos).sum()

    resumo.append({
        'coluna': coluna,
        'tipo_esperado': tipo,
        'total': len(df_tratado),
        'validos': qtd_validos,
        'invalidos_ou_nulos': qtd_invalidos,
        'percentual_valido': round(qtd_validos / len(df_tratado) * 100, 2),
        'percentual_invalido': round(qtd_invalidos / len(df_tratado) * 100, 2)
    })

df_resumo_validacao = pd.DataFrame(resumo)

df_validos = df_resumo_validacao[['coluna', 'tipo_esperado', 'total', 'validos', 'percentual_valido']]
df_invalidos = df_resumo_validacao[['coluna', 'tipo_esperado', 'total', 'invalidos_ou_nulos', 'percentual_invalido']]
```
